````python

import numpy as np 
import matplotlib.pyplot as plt 

plt.rcParams["figure.figsize"] = (10,5)
  
X = ['S1','S2','S3','S4','S5','S6','S7','S8','S9','S10','S11','S12','S13','S14','S15','S16','S17','S18','S19','S20']
Nov14 = [2.1, 2.7, 1.2, 2.1, 1.3, 1.1, 2, 1, 2, 2.2, 1.2, 1.7, 1.1, 1.9, 2.2, 2.5, 1.3, 1.7, 2.5, 2]
Nov23 = [2.1, 2.7, 1.2, 2.1, 1.3, 1.1, 2, 1, 2, 2.2, 1.2, 1.7, 1.1, 1.9, 2.2, 2.5, 1.3, 1.7, 2.5, 2]

X_axis = np.arange(len(X))  
plt.bar(X_axis - 0.2, Nov14, 0.35, label = '14-Nov-22')
plt.bar(X_axis + 0.2, Nov23, 0.35, label = '23-Nov-22')
  
plt.axis([-1, 19.8, 0, 3.2])    
plt.xticks(X_axis, X)
plt.xlabel("Statements")
plt.ylabel("Scores")
# plt.title("Number of Students in each group")
plt.legend()
plt.show()
````
