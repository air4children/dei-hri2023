````python
import numpy as np 
import matplotlib.pyplot as plt 

plt.rcParams["figure.figsize"] = (10,3)
  
X = ['S1','S2','S3','S4','S5','S6','S7','S8','S9','S10',
     'S11','S12','S13','S14','S15','S16','S17','S18','S19','S20']
Nov14 = [1.77, 1.33, 2.77, 1.77, 2.66, 3.00, 2.11, 3.00, 1.88, 1.77, 
         2.77, 2.33, 2.88, 2.11, 1.66, 1.55, 2.88, 2.33, 1.44, 1.88]
Nov23 = [2.11, 1.66, 2.77, 1.88, 2.44, 2.77, 1.88, 2.22, 1.88, 2.22, 
         2.88, 2.33, 2.66, 2.11, 2.00, 1.44, 2.77, 2.44, 2.11, 2.22]

X_axis = np.arange(len(X))  
plt.bar(X_axis - 0.2, Nov14, 0.35, label = 'L01')
plt.bar(X_axis + 0.2, Nov23, 0.35, label = 'L04')
  
plt.axis([-1, 19.8, 0, 3.2])    
plt.xticks(X_axis, X)
plt.xlabel("Statements")
plt.ylabel("Average scores")
# plt.title("Number of Students in each group")
plt.legend()
# plt.grid()
plt.axhline(y = 1, color = 'b', linestyle = '--', linewidth=0.1)
plt.axhline(y = 1.5, color = 'b', linestyle = '--', linewidth=0.25)
plt.axhline(y = 2, color = 'b', linestyle = '--', linewidth=0.1)
plt.axhline(y = 2.5, color = 'b', linestyle = '--', linewidth=0.25)
plt.axhline(y = 3, color = 'b', linestyle = '--', linewidth=0.1)
plt.show()
````
