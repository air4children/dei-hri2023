````python
import plot_likert
import pandas as pd
import numpy as np
````

````python
# load tha data as a data frame
tables = pd.read_html('http://www.mishou.be/2021/10/04/pythonr-sample-data-for-data-analysis/')
df2 = tables[2]
df2
````

````python
# bef4
df2_bef4 = df2.loc[df2.classes == "bef4"]
df2_bef4.head()
````

````python
# drop three columns
df2_bef4_6 = df2_bef4.drop(columns=["id", "classes", "res3"])
df2_bef4_6.head()
````

````python
# define my selections
myscale1 = \
    ['strongly disagree',
     'disagree',
     'neutral',
     'agree',
     'strongly agree']
# create a likert plot
plot_likert.plot_likert(df2_bef4_6, myscale1, plot_percentage=True)
````

````python

rng = np.random.default_rng(seed=42)
# print(plot_likert.scales.agree) #['Strongly disagree', 'Disagree', 'Neither agree nor disagree', 'Agree', 'Strongly agree']
data = pd.DataFrame(rng.choice(plot_likert.scales.agree, (5,3)), columns=['Q1','Q2', 'Q3'])
data
````

````python
# plot_likert.plot_likert(data, plot_likert.scales.agree);

plot_likert.plot_likert(data, plot_likert.scales.agree, 
                        plot_percentage=True,
                        bar_labels=True, 
                        bar_labels_color="snow", 
                        colors=plot_likert.colors.default_with_darker_neutral,
                        figsize=(10,3))
````

````python

rng = np.random.default_rng(seed=42)
data_3 = pd.DataFrame(rng.choice(['Strongly disaqgree', 'Neither agree nor disagree', 'Strongly agree'], (5,4)), 
                    columns=['Q1','Q2', 'Q3', 'Q4'])
data_3
````

````python
plot_likert.plot_likert(data_3, ['Strongly disaqgree', 'Neither agree nor disagree', 'Strongly agree'], 
                        plot_percentage=True,
                        bar_labels=True, 
                        bar_labels_color="snow", 
                        #colors=plot_likert.colors.default_with_darker_neutral,
                        colors=plot_likert.colors.likert4,
                        figsize=(10,3))

````
